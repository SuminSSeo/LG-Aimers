```python
# 각 열의 결측치 비율 계산 =>gpt 코드
missing_data_percentage = df_cleaned.isnull().mean() * 100
print(missing_data_percentage)
#  isnull() : 데이터프레임에서 결측치(NaN)을 True, 결측치가 아닌 값을 False로 표시하는 동일한 크기의 boolean 데이터프레임 생성 
# mean() : 평균값을 계산하지만 여기서는 true =1, false=0으로 계산하여 각 열의 결측치 비율 계산
# mean으로 계산된 결측치 비율이 소수점으로 나오므로 퍼센트로 변환하기 위해 100 곱해줌
# ...으로 생략되어있기 때문에 아래 두 코드 사용함 
```

```python
# 'Normal'과 'Abnormal' 값을 0과 1로 변환 (loc 사용)
# 문자열로 변환하고 공백 제거 및 대소문자 변환 후, 값을 0과 1로 변환
# map을 사용하여 변환
#df_cleaned['target'] = df_cleaned['target'].map({'Normal': 0, 'Abnormal': 1})
```

```python
# 'normal'과 'abnormal'을 각각 0과 1로 변환
# chat gpt 코드 => 자꾸 abnormal은 변경이 안되고, normal만 0으로 변경됨(왜?), 사용한 method는 map을 사용함
# 원본 데이터프레임의 복사본 생성
df_cleaned = df_cleaned.copy()

# target 열의 데이터 타입을 문자열로 변환하고, 공백 제거 및 소문자로 변환
df_cleaned['target'] = df_cleaned['target'].astype(str).str.strip().str.lower()

df_cleaned['target'] = df_cleaned['target'].replace({'normal': 0, 'abnormal': 1})
```

```python
# ok와 NaN만 있는 열은 NaN을 0으로 교환 
# HEAD NORMAL COORDINATE X AXIS(Stage1) Judge Value_Fill2,HEAD NORMAL COORDINATE X AXIS(Stage1) Judge Value_Fill1, GMES_ORIGIN_INSP_JUDGE_CODE Judge Value_AutoClave
# HEAD NORMAL COORDINATE X AXIS(Stage1) Judge Value_Dam, GMES_ORIGIN_INSP_JUDGE_CODE Collect Result_AutoClave
   
# 변경할 열 목록
change_columns = [
    'HEAD NORMAL COORDINATE X AXIS(Stage1) Collect Result_Dam',
    'HEAD NORMAL COORDINATE X AXIS(Stage1) Judge Value_Dam',
    'GMES_ORIGIN_INSP_JUDGE_CODE Collect Result_AutoClave',
    'GMES_ORIGIN_INSP_JUDGE_CODE Judge Value_AutoClave',
    'HEAD NORMAL COORDINATE X AXIS(Stage1) Collect Result_Fill1',
    'HEAD NORMAL COORDINATE X AXIS(Stage1) Judge Value_Fill1',
    'HEAD NORMAL COORDINATE X AXIS(Stage1) Collect Result_Fill2',
    'HEAD NORMAL COORDINATE X AXIS(Stage1) Judge Value_Fill2'
]
# NaN을 0으로, 'OK'를 1로 변환
for col in change_columns:
    df_cleaned.loc[:, col] = df_cleaned[col].apply(lambda x: 1 if x == 'OK' else (0 if pd.isna(x) else x))
```
